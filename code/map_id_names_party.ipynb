```python
import pandas as pd

data = pd.read_csv('C:\\Users\\krist\\Documents\\Uni\\RWTH\\WS23\\Master_project\\Data\\combined_DE_politician_twitter_timelines_2016-01-01_to_2023-02-11_clean.txt', sep = ",")
mdbs = pd.read_csv('C:\\Users\\krist\\Documents\\Uni\\RWTH\\WS23\\Master_project\\Data\\DE_politician_twitter_users.csv')
```

```python
mdbs['author_id'].head()
```

```python
data.head()
```

```python
replace = {}
for index,row in mdbs.iterrows():
    replace[row['author_id']] = row['name']
```

```python
data_with_authors = data.copy()
data_with_authors['name'] = data_with_authors['author_id'].copy()
data_with_authors.replace(to_replace={"name":replace},inplace=True)
```

```python
data_with_authors.head()
```

```python
replace_party = {}
for index,row in mdbs.iterrows():
    replace_party[row['author_id']] = row['party']

data_with_authors['party'] = data_with_authors['author_id'].copy()
data_with_authors = data_with_authors.replace(to_replace={'party':replace_party})
```
