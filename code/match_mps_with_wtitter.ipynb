```python
import pandas as pd
import datetime
```

```python
legislative_period_19_begin = datetime.date(2017, 10, 24)
legislative_period_19_end = datetime.date(2021, 10, 26)
```

```python
twitter_users = pd.read_csv('C:\\Users\\krist\\Documents\\Uni\\RWTH\\WS23\\Master_project\\Data\\DE_politician_twitter_users.csv')
mps = pd.read_excel('C:\\Users\\krist\\Documents\\Uni\\RWTH\\WS23\\Master_project\\Data\\MPs_WP19.xlsx',index_col=0)
```

```python
#cast twitter users 'created_at' to date
twitter_users['created_at'] = pd.to_datetime(twitter_users['created_at']).dt.date
twitter_users = twitter_users[twitter_users['created_at'] < legislative_period_19_end]
```

```python
twitter_users['start'] = [pd.NA]*len(twitter_users)
twitter_users['end'] = [pd.NA]*len(twitter_users)
```
