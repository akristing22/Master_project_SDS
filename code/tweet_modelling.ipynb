```python
import pandas as pd
from bertopic import BERTopic
from umap import UMAP

from bertopic.representation import KeyBERTInspired
from bertopic.vectorizers import ClassTfidfTransformer

```

```python
tweet_data = pd.read_csv('../data/tweets_from_mps_19WP_cleaned.csv')
```

```python
tweet_data.info()
tweet_data.head()
```

```python
print("starting cell")
cut_data = tweet_data.text.astype(str)
umap_model = UMAP(n_neighbors=15, min_dist=0.2, metric='cosine', low_memory=False,n_components=2)
representation_model = KeyBERTInspired()
ctfidf_model = ClassTfidfTransformer(bm25_weighting=True, reduce_frequent_words=True)


topic_model = BERTopic(umap_model = umap_model , language= "multilingual", 
                       min_topic_size=300, representation_model=representation_model,
                      ctfidf_model = ctfidf_model)
topics, probs = topic_model.fit_transform(cut_data)
```

```python
topic_model.save("tweet_model", serialization="pickle")
topic_info = topic_model.get_topic_info()
topic_info
```
