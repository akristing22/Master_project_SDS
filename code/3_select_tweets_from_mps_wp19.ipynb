```python
import pandas as pd
import datetime as dt
```

```python
## REPLACE WITH YOUR PATH
data = pd.read_csv('YOURPATH\\combined_DE_politician_twitter_timelines_2016-01-01_to_2023-02-11_clean.txt', sep = ",")
mps = pd.read_csv('..\\data\\twitter_mps_wp19.csv', sep = ",")
```

```python
#cast dates to dateime
mps['start'] = pd.to_datetime(mps['start'],utc=True)
mps['end'] = pd.to_datetime(mps['end'],utc=True)
data['created_at'] = pd.to_datetime(data['created_at'])

legislative_period_19_begin = pd.Timestamp("2017-10-24", tz="UTC").replace(microsecond=0)
legislative_period_19_end = pd.Timestamp("2021-10-26", tz="UTC").replace(microsecond=0)
```

```python
#filter tweets only in 19th legislative period
data = data[data['created_at'] > legislative_period_19_begin]
data = data[data['created_at'] < legislative_period_19_end]
```

```python
#match the author ids of tweets with names, party, start and end dates of mps
replace = {}
replace_party = {}
starts = {}
ends = {}
for index,row in mps.iterrows():
    replace[row['author_id']] = row['name']
    replace_party[row['author_id']] = row['party']
    starts[row['author_id']] = row['start']
    ends[row['author_id']] = row['end']
```

```python
#add columns and fill values for names, party, start and end dates of mps
data_with_authors = data.copy()
data_with_authors['name'] = data_with_authors['author_id'].copy()
data_with_authors['party'] = data_with_authors['author_id'].copy()
data_with_authors['start'] = data_with_authors['author_id'].copy()
data_with_authors['end'] = data_with_authors['author_id'].copy()
data_with_authors.replace(to_replace={'party':replace_party}, inplace=True)
data_with_authors.replace(to_replace={"name":replace},inplace=True)
data_with_authors.replace(to_replace={"start":starts},inplace=True)
data_with_authors.replace(to_replace={"end":ends},inplace=True)
```

```python
data_with_authors.tail()
```

```python
#only keep tweets which were matched with an mp of the 19th legislative period
data_with_authors = data_with_authors[data_with_authors['start'] != data_with_authors['author_id']]
#only keep tweets which were created during the mp's term
data_with_authors = data_with_authors[data_with_authors['start']<data_with_authors['created_at']]
data_with_authors = data_with_authors[data_with_authors['end']>data_with_authors['created_at']]
```
