```python
from sklearn.metrics import cohen_kappa_score
import pandas as pd
import numpy as np
```

```python
mapping_df = pd.read_csv("../data/topic_mapping_03_law_kris_jona.csv", encoding="ISO-8859-1", on_bad_lines='skip', index_col=0, delimiter=';')
mapping_df
```

```python
from sklearn.metrics import cohen_kappa_score

# Assuming the columns 'kris', 'jona', and 'lawrence' exist in the dataframe df
reader1_annotations = mapping_df['Kris']
reader2_annotations = mapping_df['Jona']
reader3_annotations = mapping_df['Lawrence']

# Calculate the Cohen's kappa score
kappa_score_1_2 = cohen_kappa_score(reader1_annotations, reader2_annotations)
kappa_score_1_3 = cohen_kappa_score(reader1_annotations, reader3_annotations)
kappa_score_2_3 = cohen_kappa_score(reader2_annotations, reader3_annotations)

print(kappa_score_1_2, kappa_score_1_3, kappa_score_2_3)

np.mean([kappa_score_1_2, kappa_score_1_3, kappa_score_2_3], axis=0)
```

```python
from statsmodels.stats import inter_rater as irr


# Assuming the columns 'kris', 'jona', and 'lawrence' exist in the dataframe df
reader1_annotations = mapping_df['Kris'].values
reader2_annotations = mapping_df['Jona'].values
reader3_annotations = mapping_df['Lawrence'].values

table = np.array([reader1_annotations, reader2_annotations, reader3_annotations]).T
agg = irr.aggregate_raters(table) # returns a tuple (data, categories)
irr.fleiss_kappa(agg[0], method='fleiss')
```

```python
twitter_df = pd.read_excel("../data/twitter_model_Super_all.xlsx")
twitter_df = twitter_df[twitter_df['Topic'].notna()]
twitter_df
```
