```python
import pandas as pd
from bertopic import BERTopic
```

```python
data = pd.read_json('data/speeches.jsonl', lines=True)

data.head()
```

```python
data.describe()
```

```python
text = data['text']

text.head()

# for testinng reduce the size of the data
text = text[:1000]
```

```python
## cleanup of tweets
import re
import nltk
from nltk.corpus import stopwords

def clean_text(text):
    text = re.sub(r'@[A-Za-z0-9]+', '', text) # Remove @mentions
    text = re.sub(r'#', '', text) # Remove '#' hash tag
    text = re.sub(r'RT[\s]+', '', text) # Remove RT
    text = re.sub(r'https?:\/\/\S+', '', text) # Remove hyperlink
    text = re.sub(r'\n', '', text) # Remove newline
    text = re.sub(r'\r', '', text) # Remove newline
    text = re.sub(r'&amp;', '', text) # Remove newline
    text = re.sub(r'&gt;', '', text) # Remove newline
    text = re.sub(r'&lt;', '', text) # Remove newline    
    return text

# Function to remove the first sentence
def remove_first_sentence(text):
    sentences = text.split('. ')  # Split by period and space
    if len(sentences) > 1:
        return '. '.join(sentences[1:])  # Join all sentences except the first one
    else:
        return text  # If there's only one sentence or no sentence, return the original text

# Apply the function to the 'text' column



data['text'] = data['text'].apply(clean_text)
data['text'] = data['text'].apply(remove_first_sentence)
data.head()

reducedData = data.text
reducedData.head()
```

```python
topic_model = BERTopic(language="german", min_topic_size=20)
topics, probs = topic_model.fit_transform(reducedData)
```
